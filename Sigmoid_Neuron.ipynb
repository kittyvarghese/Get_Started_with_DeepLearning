```
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```
def sigmoid(x,w,b):
  y = 1/(1+(np.exp(-(w*x+b))))
  return y
```

```
w = 1#@param {type:"slider", min:-10, max:10, step:0.1}
x = np.linspace(-2,2,100)
b = 0
y = sigmoid(x,w,b)
plt.plot(x,y)
plt.show()
```

3d dimension dat

```
def sigmoid_d(x1,w1,x2,w2,b):
  return 1/(1+np.exp(-(w1*x1+w2*x2+b)))
```

```
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
b=0 #@param {type:"slider", min:-10, max:10, step:1}
w1=0.5 #@param {type:"slider", min:-10, max:10, step:1}
w2=0.5 #@param {type:"slider", min:-10, max:10, step:1}
X1, X2 = np.meshgrid(x1, x2)
y = sigmoid_d(X1,w1,X2,w2,b)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X1,X2, y, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
#ax.contour3D(X1, X2, y, 450, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

**Plotting loss function**

Lets create a dataset

```
X = np.random.random(50)*20-10
W = 0.5
B = 0.2
Y = sigmoid(X,W,B)
plt.plot(X,Y,'*')
plt.show()
```

Now, lets create a algorithm for loss

```
def calculate_loss(X,Y,w_est,b_est):
  loss = 0
  for x,y in zip(X,Y):
    loss =+ (y- sigmoid(x,w_est,b_est))**2
    return loss
```

```
W = np.linspace(-1,1,100)
B = np.linspace(-1,1,100)
WW,BB = np.meshgrid(W,B)

Loss = np.zeros(WW.shape)
```

```
for i in range(WW.shape[0]):
  for j in range(WW.shape[1]):
    Loss[i,j] = calculate_loss(X,Y,WW[i,j],BB[i,j])
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(WW,BB, Loss, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
#ax.contour3D(X1, X2, y, 450, cmap='viridis')
ax.set_xlabel('Weights')
ax.set_ylabel('Bias')
ax.set_zlabel('Loss');
```
